```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['MonthYear'] = df['Order Date'].dt.to_period('M')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df.head()
```

```python
def assign_quarter(Month):
    if Month <= 3:
        return 'Q1'
    elif Month >=4 and Month <= 6:
        return 'Q2'
    elif Month >= 7 and Month <= 9:
        return 'Q3'
    elif Month >= 9 and Month <= 12:
        return 'Q4'
    else:
        return np.nan
df['Quarter'] = df['Month'].apply(assign_quarter)
df.head(300)
```

### Let's find out who our most valuable customers are!

#### The most valuable customers are those who are the most profitable for a company. These customers buy more or higher-value products than the average customer.

```python
top_cust_df = df.groupby(['Customer Name']).sum().sort_values('Sales', ascending=False).head(20)
top_cust_df = top_cust_df[['Sales']].round(2)
top_cust_df.reset_index(inplace=True)
```

```python
plt.figure(figsize = (15,5))
plt.title("Most Valuable Customers (2011-2014)", fontsize=18)
plt.bar(top_cust_df['Customer Name'], top_cust_df['Sales'], color='y', edgecolor='b', linewidth=1)
plt.xlabel('Customers', fontsize=15)
plt.ylabel('Revenue', fontsize=15)
plt.xticks(fontsize=12, rotation=60)
plt.yticks(fontsize=12)
for k, v in top_cust_df['Sales'].items():
    plt.text(k,v-7000, '$' + str(v), fontsize=12, rotation=90, color='k', horizontalalignment='center')
```

### Let's figure out which states generated the highest revenue!

#### Here are the Top 10 States which generated the highest revenue

```python
top_states_df = df.groupby(['State']).sum().sort_values('Sales', ascending=False).head(20)
top_states_df = top_states_df[['Sales']].round(2)
top_states_df.reset_index(inplace=True)
```

```python
plt.figure(figsize=(15,5))
plt.title("States with Highest Revenue (2011 - 2014)", fontsize=18)
plt.bar(top_states_df['State'], top_states_df['Sales'], color='y', edgecolor='r', linewidth=1)
plt.xlabel("States", fontsize=15)
plt.ylabel("Revenue", fontsize=15)
plt.xticks(fontsize=12, rotation=65)
plt.yticks(fontsize=12)
for k,v in top_states_df['Sales'].items():
    if v > 400000:
        plt.text(k,v-150000, '$' + str(v), fontsize=12, rotation=90, color='k', horizontalalignment='center')
    else:
        plt.text(k,v+15000, '$' + str(v), fontsize=12, rotation=90, color='k', horizontalalignment='center')
```

### Technique Report

- After experimenting with x axis label rotations, I found that 65 degrees provides the most readability without the labels wandering too far outside of their bars

### Let's look at the revenue generated by each category

```python
top_cat_df = df.groupby(df['Category']).sum().sort_values(['Sales'], ascending=False)
top_cat_df = top_cat_df[['Sales']]
total_rev_cat = top_cat_df['Sales'].sum()
total_rev_cat = str(int(total_rev_cat))
total_rev_cat = '$' + total_rev_cat
top_cat_df.reset_index(inplace=True)
```

```python
plt.rcParams['figure.figsize'] = (13,5)
plt.rcParams['font.size'] = 12.0
plt.rcParams['font.weight'] = 6
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total)/100.0)
        return '${v:d}'.format(v=val)
    return my_format
colors = ['b', 'g', 'r']
explode = (0.05, 0.05, 0.05)
fig1, ax1 = plt.subplots()
ax1.pie(top_cat_df['Sales'], colors = colors, labels = top_cat_df['Category'], 
        autopct = autopct_format(top_cat_df['Sales']), startangle=90, explode = explode)
center_circle = plt.Circle((0,0), 0.83, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
ax1.axis('equal')
label = ax1.annotate('Total Revenue \n' + str(total_rev_cat), color = 'blue', xy=(0,0), fontsize=12, ha='center')
plt.tight_layout()
plt.show()
```

### Result:
- Technology had the highest sales of the three categories with $827,455

- Total Revenue was $2,261,536

### Result:
- Canon imageCLASS 2200 Advanced Copier was the highest grossing product with $61,599

- Total Revenue from the top 8 highest grossing products was $209,624

### Technique Report:
- Here I initially chose colors that were next to each other in the color list but they ended up looking too similar in the chart. Then I tried skipping every other color in the list and the result was a much more noticeable distinction between colors.

### Let's look at Revenue by Subcategory

```python
subcat_df = df.groupby(['Category', 'Sub-Category']).sum().sort_values('Sales', ascending=False).head(10)
subcat_df = subcat_df[['Sales']].astype(int)
subcat_df = subcat_df.sort_values(['Category'])
subcat_df.reset_index(inplace=True)
subcat_df1 = subcat_df.groupby(['Category']).sum()
subcat_df1.reset_index(inplace=True)
```

```python
plt.rcParams['figure.figsize'] = (15,10)
fig, ax = plt.subplots()
ax.axis('equal')
width=0.1
outer_colors = ['goldenrod', 'gold', 'khaki']
inner_colors = ['darkorange', 'orange', 'burlywood', 'navajowhite', 'palegoldenrod', 
                'darkkhaki', 'yellow', 'bisque', 'wheat', 'lemonchiffon']
pie = ax.pie(subcat_df1['Sales'], radius=1, labels=subcat_df1['Category'], colors = outer_colors, wedgeprops=dict(edgecolor='w'))
pie2 = ax.pie(subcat_df['Sales'], radius=1-width, labels=subcat_df['Sub-Category'], autopct = autopct_format(subcat_df['Sales']), labeldistance=0.7, colors=inner_colors, wedgeprops=dict(edgecolor='w'), pctdistance=0.53, rotatelabels=True)

fraction_text_list = pie2[2]
for text in fraction_text_list:
    text.set_rotation(0)
center_circle = plt.Circle((0,0),0.6,fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
ax1.axis('equal')
plt.tight_layout()
plt.show()
```

### Result:
- Phones have the highest Subcategory sales of $327,782
